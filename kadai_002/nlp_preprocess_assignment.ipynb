```python
from bs4 import BeautifulSoup
from urllib import request
import re
```

```python
url = "https://www.aozora.gr.jp/cards/000148/files/2371_13943.html"
response = request.urlopen(url)
soup = BeautifulSoup(response, "html.parser")
response.close()

print("HTML取得成功")
```

```python
main_div = soup.find("div", class_="main_text")
print(main_div is not None)
```

```python
# ルビ削除
for tag in main_div.find_all(["rt", "rp"]):
    tag.decompose()

# テキストだけ抽出
main_text = main_div.get_text()

# 改行・全角空白を削除
main_text = re.sub(r"[\u3000\r\n]+", "", main_text)

print(main_text[:300])
```

```python
first_sentence = main_text.split("。")[0] + "。"
print(first_sentence)
```

```python
# first_sentence を見ながら、人が区切る
split_text_list = [
    "近頃", "は", "〜〜", "。"
]

print(split_text_list)
```

```python
stopwords_text = """
は
の
です
ます
私
俺
"""

stopwords_list = [w for w in stopwords_text.split("\n") if w]

result_text_list = []
for word in split_text_list:
    if word not in stopwords_list:
        result_text_list.append(word)

print("分割前:", split_text_list)
print("除去後:", result_text_list)
```
